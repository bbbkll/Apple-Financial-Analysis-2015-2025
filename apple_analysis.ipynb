# Comprehensive Business Analysis: Big Tech Revenue (2019-2023)

## 1. Project Overview and Methodology
This project analyzes the financial performance of the top five global technology companies. The analysis utilizes an offline, relative-path dataset to guarantee full reproducibility without relying on unstable external APIs.
- **Data Source**: Official corporate 10-K filings.
- **Analysis Dimensions**: Growth Trends, Aggregate Market Expansion, and Market Share.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Data Loading via Relative Path
Loading the localized CSV file using a relative path to ensure the notebook runs seamlessly across different environments.

```python
# Load data using relative path from the current directory
file_path = './tech_revenue_data.csv'
df = pd.read_csv(file_path, index_col='Year')

# Define global background color for academic consistency
bg_color = (249/255, 240/255, 227/255)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

df.head()
```

## 3. Visual Analysis 1: Individual Revenue Growth Trends
A multi-line chart illustrating the individual revenue trajectories of each company over the five-year period.

```python
fig1, ax1 = plt.subplots(figsize=(12, 6))
fig1.patch.set_facecolor(bg_color)
ax1.set_facecolor(bg_color)

for i, company in enumerate(df.columns):
    ax1.plot(df.index, df[company], marker='o', linewidth=2.5, label=company, color=colors[i])

ax1.set_title('Revenue Trajectories of Tech Giants (Billions USD)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Financial Year', fontsize=12)
ax1.set_ylabel('Revenue (Billions USD)', fontsize=12)
ax1.set_xticks(df.index)
ax1.legend(title='Company', loc='upper left')
ax1.grid(True, color='gray', alpha=0.2, linestyle='--')

plt.tight_layout()
plt.show()
```

## 4. Visual Analysis 2: Aggregate Market Scale (Stacked Bar Chart)
This stacked bar chart demonstrates the cumulative revenue of all five companies, highlighting the overall expansion of the tech sector ecosystem.

```python
fig2, ax2 = plt.subplots(figsize=(12, 6))
fig2.patch.set_facecolor(bg_color)
ax2.set_facecolor(bg_color)

bottom = np.zeros(len(df))

for i, company in enumerate(df.columns):
    ax2.bar(df.index, df[company], bottom=bottom, label=company, color=colors[i], alpha=0.85)
    bottom += df[company].values

ax2.set_title('Cumulative Annual Revenue of Top 5 Tech Firms', fontsize=14, fontweight='bold')
ax2.set_xlabel('Financial Year', fontsize=12)
ax2.set_ylabel('Total Revenue (Billions USD)', fontsize=12)
ax2.set_xticks(df.index)
ax2.legend(title='Company', loc='upper left')
ax2.grid(axis='y', color='gray', alpha=0.2, linestyle='--')

plt.tight_layout()
plt.show()
```

## 5. Visual Analysis 3: Market Share Distribution in 2023 (Pie Chart)
A cross-sectional analysis of the most recent financial year (2023) to visualize the proportionate market dominance among these five leaders.

```python
fig3, ax3 = plt.subplots(figsize=(8, 8))
fig3.patch.set_facecolor(bg_color)
ax3.set_facecolor(bg_color)

revenue_2023 = df.loc[2023]
explode = (0.05, 0, 0, 0.05, 0)  # Highlight top performers

ax3.pie(revenue_2023, labels=revenue_2023.index, autopct='%1.1f%%', startangle=140, 
        colors=colors, explode=explode, textprops={'fontsize': 12})

ax3.set_title('Revenue Distribution Share in 2023', fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

## 6. Conclusion
The multi-dimensional analysis successfully utilizes local data reading via relative path. The visualizations confirm significant aggregate growth in the tech sector from 2019 to 2023, with distinct distribution of market share among the leading entities.
